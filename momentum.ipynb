```python
%%manim -qm -v WARNING CreateCircle

class MovingRectangle(Rectangle):
    def __init__(self, direction, velocity: float, **kwargs):
        super().__init__(**kwargs)
        self.direction = direction
        self.velocity = velocity
 
class CreateCircle(Scene):
    def construct(self):
        rect = MovingRectangle(LEFT,1.0,width=4.0,height=3.0)
        self.add(rect)
        self.add(Line(start=LEFT*(config.frame_width/2),end=RIGHT*config.frame_width/2))
        print(config.frame_width-4)
        self.play(rect.animate.shift(RIGHT),run_time=3)
```
